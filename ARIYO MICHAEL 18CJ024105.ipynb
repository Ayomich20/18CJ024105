```python
African_Countries = ['Algeria',
                        'Angola',
                        'Benin',
                        'Botswana',
                        'Burkina Faso',
                        'Burundi',
                        'Cameroon',
                        'Cape Verde',
                        'Central African Republic',
                        'Chad',
                        'Comoros',
                        "Côte d'Ivoire",
                        'Democratic Republic of the Congo',
                        'Djibouti',
                        'Egypt',
                        'Equatorial Guinea',
                        'Eritrea',
                        'Eswatini',
                        'Ethiopia',
                        'Gabon',
                        'Gambia',
                        'Ghana',
                        'Guinea',
                        'Guinea-Bissau',
                        'Kenya',
                        'Lesotho',
                        'Liberia',
                        'Libyan Arab Jamahiriya',
                        'Madagascar',
                        'Malawi',
                        'Mali',
                        'Mauritania',
                        'Mauritius',
                        'Morocco',
                        'Mozambique',
                        'Namibia',
                        'Niger',
                        'Nigeria',
                        'Rwanda',
                        'Sao Tome and Principe',
                        'Senegal',
                        'Seychelles',
                        'Sierra Leone',
                        'Somalia',
                        'South Africa'
                        'South Sudan',
                        'Sudan',
                        'Tanzania',
                        'Togo',
                        'Tunisia',
                        'Uganda'
                        'Zambia',
                        'Zimbabwe', ]
df = pd.read_csv("cat.csv")

Production = df[df['Area'].isin(African_Countries)]

x = Production['Area']
y = Production['Value']
Production.set_index('Area', inplace=True)


for idx,year in enumerate(list(range(2010, 2020))):
        fig, ax = plt.subplots(1)
        Production[(Production['Year'] == year) & (Production['Item'] == 'Cattle')]['Value'].plot.barh(ax=ax, y='Value', title=year, figsize=(12, 12))

```

```python
mean_vals = []
for year in range(2010,2019) :
    mean_vals.append(df[df['Year'] == year]['Value'].mean())
    
plt.plot(range(2010,2019), mean_vals)
plt.title('Mean for Meat Stock in Africa from 2010-2019')
plt.xlabel('Year')
plt.ylabel('Value')
```

```python
stdev_vals = []
for year in range(2010,2019):
    stdev_vals.append(df[df['Year'] == year]['Value'].var()**0.5)
    
plt.plot(range(2010,2019), stdev_vals)
plt.title('Standard Deviation for Meat Stock in Africa from 2010-2019')
plt.xlabel('Year')
plt.ylabel('Value')
```

```python
median_vals = []
for year in range(2010,2019) :
    median_vals.append(df[df['Year'] == year]['Value'].median())
    
plt.plot(range(2010,2019), median_vals)
plt.title('Median for Meat Stock in Africa from 2010-2019')
plt.xlabel('Year')
plt.ylabel('Value')
```

```python
fig, axes  = plt.subplots(1,9, figsize=(10,10), sharey=True)
fig.suptitle('Boxplots from 2010 to 2018')
axes[0].set_ylabel('Stocks')

for idx, year in enumerate(range(2010,2019)):
    sns.boxplot(ax = axes[idx], y = df[(df['Year'] == year) & (df['Element'] == 'Stocks')]['Value'])
    axes[idx].set_title(year)
    if idx > 0: axes[idx].set_ylabel('')
```
